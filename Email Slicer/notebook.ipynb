```python
def emailProcess(email):
    email_username = email[0:email.index("@")]
    email_domain = email[email.index("@")+1:]
    return [email_username, email_domain]

def prinMsg(email_username, email_domain):
    print(f"Username {email_username} va Domain {email_domain}")

def main():
    email = input("Nhap dia chi email vao:").strip()
    email_username, email_domain =  emailProcess(email)
    prinMsg(email_username, email_domain)
    
if __name__ == "__main__":
    main()
```
